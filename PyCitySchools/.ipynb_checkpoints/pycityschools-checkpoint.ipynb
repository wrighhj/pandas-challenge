```python
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school = pd.read_csv(school_data_to_load)
student = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
pycity = pd.merge(school, student, how="left", on=["school_name", "school_name"])
pycity
```

```python
pycity_name = pycity.rename(columns={"School ID": "school_id", "school_name": "school", "Student ID": "student_id"})
pycity_clean = pycity_name.dropna(how="any")
pycity_clean
```

```python
### District Summary

# * Create a high level snapshot (in table form) of the district's key metrics, including:
#   * Total Schools
#   * Total Students
#   * Total Budget
#   * Average Math Score
#   * Average Reading Score
#   * % Passing Math (The percentage of students that passed math.)
#   * % Passing Reading (The percentage of students that passed reading.)
#   * % Overall Passing (The percentage of students that passed math **and** reading.)

# unique schools
pycity_clean["school"].nunique()
# unique students
pycity_clean["student_id"].nunique()
# there are 15 unique budgets because there are 15 unique schools
# pycity_clean["budget"].nunique()
# unique budgets
pycity_clean.budget.unique()
#total of unique budgets
sum(pycity_clean.budget.unique())
pycity_clean["math_score"].mean()
pycity_clean["reading_score"].mean()

math_pycity = pycity_clean.loc[pycity_clean['math_score'] >= 70, :]
perc_math_pass = (len(math_pycity["math_score"]) / len(pycity_clean["math_score"]))*100

read_pycity = pycity_clean.loc[pycity_clean['reading_score'] >= 70, :]
perc_read_pass = (len(math_pycity["reading_score"]) / len(pycity_clean["reading_score"]))*100

both_pycity = pycity_clean.loc[pycity_clean['reading_score' and 'math_score'] >= 70, :]
perc_both_pass = (len(both_pycity) / len(pycity_clean))*100

district_summary = pd.DataFrame({
    "Total Schools": [pycity_clean["school"].nunique()],
    "Total Students": [pycity_clean["student_id"].nunique()],
    "Total Budget": [sum(pycity_clean.budget.unique())],
    "Average Math Score": [pycity_clean["math_score"].mean()],
    "Average Reading Score": [pycity_clean["reading_score"].mean()],
    "% Passing Math": [perc_math_pass],
    "% Passing Reading":[perc_read_pass],
    "% Overall Passing": [perc_both_pass]
})
district_summary
```

```python
#   * School Type
school_type = school.set_index(["school_name"])["type"]
school_type = pd.DataFrame(school_type)
school_type
school_type_res = school_type.reset_index()
school_type_reset = school_type_res.rename(columns = {"school_name": "school"})
school_type_reset
```

```python
#   * Total Students
total_students = pycity_clean.groupby(["school"])["student_id"].count()
total_students = pd.DataFrame(total_students)
total_students_res = total_students.reset_index()
total_students_reset = total_students_res.rename(columns = {"student_id": "total_students"})
total_students_res
```

```python
#   * Total School Budget
school_budget = school.set_index(["school_name"])["budget"]
school_budget_res = school_budget.reset_index()
school_budget_res
school_budget_reset = school_budget_res.rename(columns={"school_name": "school"})
school_budget_reset
```

```python
#   * Per Student Budget
budget_merge = pd.merge(school_type_reset, school_budget_reset, on="school", how = "left")
school_merge = pd.merge(budget_merge, total_students_reset, on="school", how = "left")
school_merge
```

```python
#   * Per Student Budget
budget_per_student = (school_merge["budget"] / school_merge["total_students"])
budget_per_student
school_merge["budget_per_student"] = budget_per_student
school_merge
```

```python
#   * Average Math Score
ave_math = pycity_clean.groupby(['school'])['math_score'].mean()
ave_math
ave_math_res = ave_math.reset_index()
ave_math_res
school_merge_res = school_merge.reset_index() 
school_merge_res = pd.merge(school_merge_res, ave_math_res, on="school", how = "left")
school_merge_res
```

```python
#   * % Passing Math (The percentage of students that passed math.)

math_pycity = pycity_clean.loc[pycity_clean['math_score'] >= 70, :]
pas_math = math_pycity.groupby(['school'])['math_score'].count()
pas_math
pass_math = pd.DataFrame(pas_math)
pass_math
school_merge_math = pd.merge(school_merge_res, pass_math, on="school", how = "left")
smm = school_merge_math.rename(columns={"math_score_x": "ave_math_score", "math_score_y": "students_passing_math"})
smm
perc_pass_math = smm.apply(lambda row: (row.students_passing_math / row.total_students)*100, axis = 1)
perc_pass_math
smm["percent_passing_math"] = perc_pass_math
smm
```

```python
#   * Average Reading Score

ave_read = pycity_clean.groupby(['school'])['reading_score'].mean()
ave_read
ave_read_res = ave_read.reset_index()
ave_read_res
smm = pd.merge(smm, ave_read_res, on="school", how = "left")
smm
```

```python
#   * % Passing Reading (The percentage of students that passed reading.)
read_pycity = pycity_clean.loc[pycity_clean['reading_score'] >= 70, :]
pas_read = read_pycity.groupby(['school'])['reading_score'].count()
pas_read
pass_read = pd.DataFrame(pas_read)
pass_read
smr = pd.merge(smm, pass_read, on="school", how = "left")
smr
smr = smr.rename(columns={"reading_score_x": "ave_reading_score", "reading_score_y": "students_passing_reading"})
smr
perc_pass_read = smr.apply(lambda row: (row.students_passing_reading / row.total_students)*100, axis = 1)
perc_pass_read
smr["percent_passing_reading"] = perc_pass_read
smr
```

```python
#   * % Overall Passing (The percentage of students that passed math **and** reading.)

both_pycity = pycity_clean.loc[(pycity_clean['reading_score'] >=70) & (pycity_clean['math_score'] >= 70)]
both_pas = both_pycity.groupby(['school'])['math_score' and 'reading_score'].count()
both_pas
both_pass = pd.DataFrame(both_pas)
both_pass
smb = pd.merge(smr, both_pass, on="school", how = "left")
smb
smb = smb.rename(columns={"reading_score": "overall_passing"})
smb
perc_pass_both = smb.apply(lambda row: (row.overall_passing / row.total_students)*100, axis = 1)
perc_pass_both
smb["%_overall_passing"] = perc_pass_both
smb
```
